NLP: Natural Language Processing
Two sections:
Natural Language Understanding(model understanding)
Natural Language Generation(model generates new content)
Text Understanding:
Text Cleaning:
converting the text content into lowercase
removal of punctuation marks, special characters
removal of stopwords

Stopwords:Give all the data to the model , this increases the complexity. To avoid that, we can remove the stopwords.

Words that are meaningful for sentence formation but not really necessary for our model to learn are referred as stop words.
Example: a,again,against,are,and ....
Tokenization:
Converting the words into characters representing tokens
Converting the senteences into words representing  tokens
Converting the paragraphs into sentences representing tokens

Example: ["beautiful","easy","easily","walked","walk","walking"] : total words machine has to learn - 6
[beauty,easy,easy,walk,walk,walk] - total words machine has to learn - 3
Stemming: Converts the word to its root form
It removes prefix/suffix so that root form of the word is resulted.
happiness : happy
given : give

3 most popular stemmer types:
Portal Stemmer
Lancaster Stemmer
Snowball Stemmer

Stemming techniques are resulting in a word that doesn't exist
happiness: happi
given : give
easily : easiliLemmitization: Converts the word into base form. It uses the dictionary format. With the help of POS tags(parts of speech tags)
happiness: happiness
given: give

POS tags: Parts of Speech
Named Entity Recognition : (NER): Entities - you can image this as a set of entities/categories.
Oraganization
Location
Date/Time
Person
Example: "Sindhura works in 3RI technologies located at Pimple Saudagar. She has celebrated Ganesh Chaturthi on 27th August".
Sindhura,She - Person
3RI Technologies - Organization
Pimple Saudagar - Location
Date - 27th August
Vectorizer
Count Vectorizer(Bag of Words)

TFIDF

Word based vectors are generated that are provided to model for model training

```python
data = """ Data analysis is the process of inspecting, cleansing, transforming, and modeling data with the goal of discovering useful information, informing conclusions, and supporting decision-making.[1] Data analysis has multiple facets and approaches, encompassing diverse techniques under a variety of names, and is used in different business, science, and social science domains.[2] In today's business world, data analysis plays a role in making decisions more scientific and helping businesses operate more effectively. """
```

Text cleaning

```python
import re
data = data.lower()
pattern = r"[^a-z. ]"
data_new = re.sub(pattern,"",data)
data_new
```

Sentence TOkenize


```python

from nltk.tokenize import sent_tokenize
sent_tokens = sent_tokenize(data_new)
sent_tokens
```

```python
len(sent_tokens)
```

Bigrams(2) , Trigrams(3), Ngrams(n number of words)

```python
from nltk.tokenize import word_tokenize
word_tokens = word_tokenize(data_new)
word_tokens # these tokens are generated by unigrams - 1gram- 1word token is generated
```

```python
# bi_tokens = list(nltk.bigrams(unigram tokens)) => unigram tokens = word_tokenize(data)
bi_tokens = list(nltk.bigrams(word_tokens))
bi_tokens
```

```python
tri_tokens = list(nltk.trigrams(word_tokens))
tri_tokens
```

```python
n_tokens = list(nltk.ngrams(word_tokens,n=5))
n_tokens
```

Stopwords

```python
from nltk.corpus import stopwords
stop = stopwords.words('english') # mention the language
stop
```

```python
len(stop)
```

Removal of stopwords

```python
len(word_tokens)
```

```python
print(word_tokens[:50])
```

```python
filtered_words = []
for word in word_tokens:
  if word not in stop:
    filtered_words.append(word)
print(filtered_words)
```

Stemming

Porter Stemmer

```python
from nltk.stem import PorterStemmer

words = ["easily","walking","happiness","given","marked","inspecting"]

ps = PorterStemmer()

ps_words = [ps.stem(word) for word in words]
ps_words
```

Lancaster Stemmer


```python

from nltk.stem import LancasterStemmer

words = ["easily","walking","happiness","given","marked","inspecting"]

ls = LancasterStemmer()

ls_words = [ls.stem(word) for word in words]
ls_words
```

Snowball Stemmer


```python
from nltk.stem import SnowballStemmer

words = ["easily","walking","happiness","given","marked","inspecting"]

ss = SnowballStemmer(language='english')

ss_words = [ss.stem(word) for word in words]
ss_words
```

Lemmatization

```python
from nltk.stem import WordNetLemmatizer

word_lem = WordNetLemmatizer()

words =["easily","walking","happiness","given","marked","inspecting"]

lemma_words = [word_lem.lemmatize(word) for word in words] # by default, wordnet lemmatizer considers noun as default parts of speech
lemma_words
```

```python
from nltk.stem import WordNetLemmatizer

word_lem = WordNetLemmatizer()

words =["easily","walking","happiness","given","marked","inspecting"]

lemma_words = [word_lem.lemmatize(word,pos='v') for word in words] # by default, wordnet lemmatizer considers noun as default parts of speech
lemma_words
```
